# Úvod do pythonu

Pozor, python používá syntactically significant whitespace

Bloky kódu (bloky zdrojového textu) jsou určeny jejich odsazením. „Blokem kódu“ zde rozumím volání funkcí, příkazy if, cykly for, cykly while a další. Blok je zahájen odsazením (odskočením řádku vpravo) a končí předsazením (odskočením následujícího řádku vlevo). Nenajdeme zde žádné explicitní závorky nebo klíčová slova. To ale znamená, že používání whitespace má svůj význam a že je musíme užívat důsledně.

````python
def f1(x):
    if x < 0:
        x = -x
    return x
````

````python
f1(-2), f1(2)
````

````python
def f2(x):
    if x < 0:
        x = -x
        return x
````

````python
f2(-2), f2(2)
````

V prvním příkladu funkce f1 vrací absolutní hodnotu parametru x. V druhém příkladu funkce f2 pro kladné a nulové x nic nevrátí.

## Sekvence
Konečné uspořádané skupiny prvků indexovaných nezápornými čísly (řetezce jsou sekvence; seznamy a další také)

| Výraz            |Popis |
|:--------|:--------------------|
| `len(xs)` | vrací délku příslušné sekvence xs |
| `xs[i]` | vrací prvek sekvence na pozici i (smysl má i výraz typu xs[-2]) |
| `xs[i:j]` | vrací podsekvenci od indexu i po index j-1 (smysl mají i výrazy typu xs[:3], xs[-2:] apod., někde též xs[i:j:k]) |
| `zs = xs + ys` | spojování sekvencí |
| `for x in xs:` | smyčka přes všechny prvky sekvence |
| `x in xs` | zjištění výskytu prvku v sekvenci |
| `xs.index(x[,n])` | zjištění prvního umístění prvku v sekvenci (od pozice n) |
| `xs.count(x)` | počet výskytů prvku x v sekvenci xs |

### Match-case

V pythonu není switch; switch lze nahradit buď pomocí if-elif-else a nebo pomocí slovníku. Případně **od verze python 3.10 je** nový konstrukt **match-case** (Structural pattern matching), který se chová podobně jako switch -- viz [PEP634](https://peps.python.org/pep-0634/), [PEP636](https://peps.python.org/pep-0636/) nebo na [learnpython.com](https://learnpython.com/blog/python-match-case-statement/).

**Match-case** je jazykový konstrukt, který vám dovolí vytvářet komplexnější rozhodovací podmínky, které budou zapsány čitelněji, než hromadou if-else.

````python
# zjisteni verze pythonu programove
import sys
print(sys.version)
````

````python
# zakladni match-case
var = 'BI-PYT'
match var:
    case 'BI-PYT':
        print('Ahoj!')
    case 'BI-AAG':
        print('Cau!')
    case other:
        print('Nenalezeno.')
````

````python
# pokrocilejsi pripad match-case
colors = "red green blue orange yellow brown black cyan magenta white"
# colors = "black"
# colors = "red"
# str.split(): "a b c" -> ["a", "b", "c"]
match colors.split():
    case ['black']:
        print('Hodi se ke vsemu.')
    case ['red' | 'green', *colors_list] if 'black' in colors_list:
        print(f'Zacali jsme cervenou nebo zelenou a zbytek je:  {colors_list}')
    case ['red' | 'black']:
        print('Jenom ruda a cerna')
    case _:
        print('Tento pripad neznam.')
````

## Cyklus for
Pozor! V pythonu neexistuje obdoba následujícího kódu z javy

```java
for (int i = 0; i < 5; i++) {
  System.out.println(i);

```
protože for cyklus v pythonu cyklí průchodem po prvcích nějakého objektu (iterable), většinou sekvence:

Syntaxe funkce/generátoru `range`:
- od nuly do n: `range(`n`)`
- od m do n: `range(`m`,`n`)`
- od m do n s krokem k: `range(`m`,`n`,`k`)`



````python
for x in range(5):
    print(x, end='-')
    for x in range(5):
        print(x, end='')
    print() #new line
````

````python
x = list(range(10, 2))
print(x)
````

## Dynamické a statické typování

Python je dynamicky typovaný jazyk.

````python
def scitani(a, b):
    print("scitam ", end="")
    return a + b
````

**Můžete** volat s různými typy.

````python
print(scitani(1, 1))
print(scitani(1.2, 1))
print(scitani('a', 'hoj'))
````

Zároveň je python silně typovaný jazyk

````python
print(scitani(1 + "ahoj"))
````

Jak je to ale se statickým typováním? No python statické typování "podporuje"

````python
def scitani(a: int, b: int) -> int:
    print("scitam ", end="")
    return a + b
````

Problém je však v tom, že python je stále dynamicky typovaný jazyk a **statické typování nevynucuje**

````python
help(scitani)
print(scitani.__annotations__)
````

````python
print(scitani(1, 1))
print(scitani('a', 'b'))
print(scitani('a', str(1)), "protoze typ str(1) =", type(str(1)))
````

Pokud chcete nějakým způsobem ověřit, že máte typování ok, můžete použít ke kontrole použít [mypy](https://mypy.readthedocs.io/en/stable/) (`pip3 install mypy`):
