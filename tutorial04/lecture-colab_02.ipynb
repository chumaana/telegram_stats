# 1. Python a syntaxe

Python používá syntactically significant whitespace

Bloky kódu (bloky zdrojového textu) jsou určeny jejich odsazením. „Blokem kódu“ zde rozumím volání funkcí, příkazy if, cykly for, cykly while a další. Blok je zahájen odsazením (odskočením řádku vpravo) a končí předsazením (odskočením následujícího řádku vlevo). Nenajdeme zde žádné explicitní závorky nebo klíčová slova. To ale znamená, že používání whitespace má svůj význam a že je musíme užívat důsledně.

Příklad: f2(x) vrátí výsledek pouze pro záporné x

````
def f1(x):
  if x < 0:
    x = -x
  return x

def f2(x):
  if x < 0:
    x = -x
    return x

print("f1(1) =",f1(1),"; f1(-1) =",f1(-1))
print("f2(1) =",f2(1),"; f2(-1) =",f2(-1))
````

# 2. Základní typy a operátory

Ukázali jsme si základní typy pro čísla: int, float a zkusili jsme si s nimi základní operace:
```
+ sčítání
- odčítání
* násobení
** mocnění
/ dělění
// celočíselné dělění
% zbytek po celočíselném dělení
```


````
a=5 
print("Type of a is", type(a))
a+=2 # sčítání
print(a)
a-=1 # odčítání
print(a)
a*=2 # násobení
print(a)
a//=3 # celočíselné dělení
print(a)
a%=3 # zbytek po celočíselném dělění
print(a)
a/=1 # dělení - pozor výsledkem dělění je vždy float (i když vyjde celé číslo)
print(a)
print("Type of a is", type(a))
````

Řekli jsme si něco málo práci s řetězci a ukázali jsme si základní operace s nimi. Dále jsme mluvili o tom, že řetězce jsou sekvence a co z toho plyne.

````
s="Ahoj"
s+=" "+"světe!" # spojování řetězců prostřednictvím operátoru pro sčítání
print(s)
print("Type of s is", type(s))
l=len(s) # délka řetězce
print("Length of s is",l)
c=s.count("h") # počet výskytů 
print("Count of char h in '",s,"' is ",c,".",sep="")
sub=s[1:4] # substring
print("Substring 2-4 of ",s,"is",sub)
````

# 2. Podmínky, cykly

## if
Ukázali jsme si konstrukci if-elif-else a řekli jsme si, že v pythonu není case (ale že ho lze nahradit pomocí slovníku).

````
vek = int(input("Zadejte svůj věk "))

if vek<0:
  print("Chyba, zadali jste záporný věk.")
elif vek<18:
  print("Dětem nenaléváme alkohol, dejte si mléko.")
else:
  print("Dáte si pivo?")
````

## for

for se v pythonu používá pro procházení prvků sekvence. 

````
s = "Ahoj světe!"
# cyklus přes všechny prvky sekvence
for x in s:
  print(x,end="#")
````

pokud potřebujeme procházet řadu čísel v nějakém rozsahu, použijeme range. Pro úplnost: range je generátor - o generátorech budeme mluvit později.

````
s = "Ahoj světe!"
for x in range(len(s)):
  print(s[x],end="#")
````

## while

Protože for vždy prochází prvky nějakého iterovatelného objektu, nemáme v python ekvivalent následujícího kódu z javy:

```java
for (int i = 0; i < 5; i++) {
  System.out.println(i);
}
```

a jediný způsob, jak takový kód nahradit je cyklus while:

````
i=0
while i<5:
  print(i)
  i+=1
````

## vsuvka: mutable vs. immutable

Immutable (neměnné) objekty se na rozdíl od mutable objektů po založení nedají měnit. Co to znamená? Například toto (str je immutable, stejně jako n-tice):

````
s = "Ahoj světe!"

print(s[0])
s[0]='B'
````

zatímco seznamy jsou mutable:

````
s = "Ahoj světe!"
seznam = list(s)
print(seznam)
seznam[0]='B'
print(seznam)
s="".join(seznam)
print(s)
````

dále jsme si říkali, že immutable objekty se předávají hodnotou, mutable odkazem:

````
s=["A","C",["C"]]
#   ^   ^    ^
#   |   |    |- třetí prvek seznamu s je seznam, tedy mutable
#   |---|--- první a druhý prvek seznamu s jsou immutable stringy
for x in s:
  x+="x"
print(s)
````

## množiny (sets)

Množiny (sets) jsou neuspořádané kolekce hodnot.

````
# množiny
m = {1,2,3}
#množiny jsou mutable
m.update([4],{5}) # do množiny přidáváme prvky pomocí update.
print(m)
# parametr update musí být iterable
m.update(6)
````

operace s množinami

````
set1 = {1,2,3,4}
set2 = {3,4,5,6}
# průnik (prvky v obou množinách)
print(set1 & set2) # nebo set1.intersection(set2)
# sjednocení (všechny prvky z obou množin)
print(set1 | set2) # nebo set1.union(set2)
# rozdíl (prvky první množiny, ktere nejsou v druhé)
print(set1 - set2) # nebo set1.difference(set2)
# symetrický rozdíl (prvky, které jsou pouze v první a pouze v druhé)
print(set1 ^ set2) # nebo set1.symmetric_difference(set2)
````

množiny lze využít k odstranění duplicit v seznamech:

````
seznam=[1,2,3,2,3,4,1,2,3,4,5]
seznam=list(set(seznam))
print(seznam)
````

## slovníky (dictionaries)

Slovníky (dictionaries) jsou neuspořádané kolekce dvojic klíč-hodnota. Slovníky jsou mutable.

````
slovnik = {
    1 : 'Ahoj',
    2 : 'Světe',
    3 : '!'
}

# seznam klíčů ve slovníku
print(slovnik.keys())
# seznam hodnot ve slovníku
print(slovnik.values())
# seznam dvojic klíč-hodnota ve slovníku
print(slovnik.items())

# test přítomnosti klíče ve slovníku
print(1 in slovnik.keys())
#nebo
print(1 in slovnik)

#změna hodnoty ve slovníku
slovnik[3]='?'
print(slovnik)

#přidání hodnoty do slovníku
slovnik[4]='!'
print(slovnik)
````

Také jsme si ukázali, že pomocí slovníku můžeme nahradit switch-case. Například tento java kód:

```java
int day = 4;
switch (day) {
  case 1:
    System.out.println("Monday");
    break;
  case 2:
    System.out.println("Tuesday");
    break;
  case 3:
    System.out.println("Wednesday");
    break;
  case 4:
    System.out.println("Thursday");
    break;
  case 5:
    System.out.println("Friday");
    break;
  case 6:
    System.out.println("Saturday");
    break;
  case 7:
    System.out.println("Sunday");
    break;
}
// Outputs "Thursday" (day 4)
```

lze v pythonu nahradit takto:

````
# buď si můžeme udělat slovník číslo - řetězec

switch = {
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday",
    6: "Saturday",
    7: "Sunday"
}

# a zavolat funkci 
print(switch[4])


# nebo si můžeme udělat slovník funkcí
switch = {
    1: lambda: print("Monday"),
    2: lambda: print("Tuesday"),
    3: lambda: print("Wednesday"),
    4: lambda: print("Thursday"),
    5: lambda: print("Friday"),
    6: lambda: print("Saturday"),
    7: lambda: print("Sunday")
}

# a zavolat si příslušnou funkci ze slovníku
switch[4]()
````

## list comprehensions

Česky generátorová notace seznamu je velice běžnou součásti jazyka python a umožňuje stručný zápis vytvoření seznamu z jiného seznamu aplikováním funkce na všechny prvky zdrojového seznamu.

````
[x**2 for x in range(20) if x%2==0]
# číst odprosřed - základem je for x in range(10)
# vpravo je omezení; do výsledného seznamu půjdou jen sudá čísla
# vlevo co se s nimi má udělat (umocnit)
````

## generator expressions

Česky generátorové výrazy. Generátory budeme probírat později, toto berte jako poznámku k předchozímu případu. Pokud potřebujeme pouze mocniny sečíst, nemusíme si generovat celý seznam a zabírat si s ním paměť. Můžeme použít něco, co nám bude generovat jednotlivé prvky postupně až budou potřeba - generátor.

````
# mnoho lidí dělá při programování tu chybu, že seznamy používá i v situacích, kdy to není úplně vhodné, 
# např. zde je zbytečně vygenerován seznam o 100 prvcích, který nám bude zabírat paměť.
s1=[x**2 for x in range(100)]
print(type(s1))
print(sum(s1))
#pokud opravdu chceme vygenerované prvky jen sečíst, můžeme si je generovat postupně generátorem:
g=(x**2 for x in range(100))
print(type(g))
print(sum(g))
````

Obdobným způsobem lze využít generátorovou notaci i pro generování slovníku či množiny.

````
l={x:x**2 for x in range(10)}
print(type(l))
print(l)
m={x**2 for x in range(10)}
print(type(m))
print(m)
````

# 5. Funkce

A tím se pomalu dostáváme k funkcím. Zatím jsme si funkce prošli jen velmi okrajově. Řekli jsme si, že funkce se zavádí pomocí klíčového slova def:

````
def funkce(x):
  print(x)
  return x**2

print(funkce(6))
````

mluvil jsem také o tom, že python je dynamicky typovaný jazyk a co to znamená

````
# takže funkci scitani
def scitani (a,b):
    print("scitam ",end="")
    return a+b

# můžete volat s různými typy.

print(scitani(1,1))
print(scitani(1.2,1))
print(scitani('a','hoj'))

````

````
# můžeme sice nadefinovat "staticky" typovanou variantu (pomocí anotací)
def scitani_int (a:int, b:int) -> int:
    print("scitam int ",end="")
    return a+b

# ale na funkci to nebude mít vliv
print(scitani_int(1,1))
print(scitani_int(1.2,1))
print(scitani_int('a','hoj'))
````

seznámili jsme se s některým základními funkcemi: `print`, `input`, `len`, `map`, `zip`

````
# input ukládá vstup od uživatele do proměnné
vstup = input("zadej něco ")
# print píše do konzole
print(vstup)
# len počítá prvky sekvence
print(len(vstup))
````

````
# zip spojuje n seznamů
s1 = [1,2,3]
s2 = [3,4,5]
print(list(zip(s1,s2)))
````

````
def f(x):
  return x**2

# map aplikuje funkci na všechny prvky sekvence
t = (1,2,3)
list(map(f,t))
````

také jsme si ukázali, že pokud je potřeba, můžeme si udělat funkci pomocí operátoru `lambda` funkci i bez definice

````
f=lambda x: x**2
for i in range(10):
  print(f(i))
# nebo velmi běžné použití s map
list(map(lambda x: x**2, range(10)))
````

Úplně poslední věc, kterou bych chtěl zopakovat je hvězdičkový operátor, který slouží k "rozbalení" sekvence a předání jednotlivých prvků sekvence funkci.

````
seznam = ["Ahoj", "světe", "!"]

print(seznam)
# pomocí * předám prvky seznamu jako jednotlivé parametry 
print(*seznam)
````
